```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Reemplaza 'tu_archivo.csv' con la ruta de tu archivo CSV
archivo_csv = '/content/historicoIEX_aapl.csv'
```

```python
# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv(archivo_csv)

# Verificar las primeras filas del DataFrame
print(df.head())

# Calcular la media y la desviación estándar
media = df.mean()
desviacion_estandar = df.std()

# Imprimir los resultados
print("\nMedia:")
print(media)

print("\nDesviación Estándar:")
print(desviacion_estandar)
```

```python
df.columns
```

```python


# Asegurarse de que la columna 'Fecha' exista en el DataFrame
if 'Fecha' in df.columns:
    # Configurar el índice como la columna 'Fecha'
    df.set_index('Fecha', inplace=True)

    # Cambiar el formato de la fecha
    df.index = pd.to_datetime(df.index).strftime('%d-%m')  # Convertir el índice a formato de fecha y luego formatear

    # Aquí puedes realizar otras operaciones si es necesario

else:
    print("La columna 'Fecha' no existe en el DataFrame.")
```

```python
# Configurar el índice como la columna 'Fecha'
#df.set_index('Fecha', inplace=True)

# Crear un gráfico de líneas para las cotizaciones diarias
plt.figure(figsize=(12, 6))
for columna in df.columns:
    plt.plot(df.index, df[columna], label=columna)

plt.title("Cotizaciones Diarias de Acciones de Apple")
plt.xlabel("Fecha - Año 2023")
plt.ylabel("Cotización")
plt.legend()
plt.xticks(rotation=60)  # Rotar etiquetas del eje x para mayor legibilidad
plt.show()
```

```python


# Graficar los valores de cierre por día
plt.figure(figsize=(12, 6))
plt.plot(df['Precio de Cierre'], marker='o', linestyle='-')
plt.title('Variación del Precio de Cierre de las acciones de Apple')
plt.xlabel('Fecha - Año 2023')
plt.ylabel('Precio de Cierre')
plt.xticks(rotation=60)  # Rotar etiquetas del eje x para mayor legibilidad
plt.grid(True)
plt.show()
```

```python
# Graficar un histograma del volumen
plt.figure(figsize=(12, 6))
plt.hist(df['Volumen'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución del Volumen de Transacciones de las acciones de Apple')
plt.xlabel('Volumen')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.show()
```
