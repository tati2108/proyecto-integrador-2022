```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Subir el archivo CSV desde tu máquina local
uploaded = files.upload()

# Obtener el nombre del archivo
archivo_csv = list(uploaded.keys())[0]

# Intenta cargar el archivo CSV con el codec 'latin1'
try:
    # Cargar el archivo CSV en un DataFrame
    df = pd.read_csv('/content/historicoYahooFinance_apple_20231111111200 (1).csv', encoding='latin1')

    # Eliminar comas y convertir columnas numéricas al tipo adecuado
    df['Abrir'] = df['Abrir'].str.replace(',', '.').astype(float)
    df['Máx'] = df['Máx'].str.replace(',', '.').astype(float)
    df['Mín'] = df['Mín'].str.replace(',', '.').astype(float)
    df['Cierre'] = df['Cierre'].str.replace(',', '.').astype(float)
    df['Cierre Ajustado'] = df['Cierre Ajustado'].str.replace(',', '.').astype(float)
    df['Volumen'] = df['Volumen'].str.replace('.', '').astype(int)



    # Convertir la columna 'Fecha' al tipo de dato datetime
    df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')  # Usa 'coerce' para manejar valores no válidos

    # Configurar el índice como la columna 'Fecha'
    df.set_index('Fecha', inplace=True)

    # Plotear los precios de cierre a lo largo del tiempo
    plt.figure(figsize=(14, 7))
    plt.plot(df.index, df['Cierre'], label='Cierre', color='blue')
    plt.title('Cotizaciones de Acciones de Apple en la Bolsa de Madrid')
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre')
    plt.legend()
    plt.show()

except UnicodeDecodeError:
    print("Error: No se pudo decodificar el archivo CSV. Verifica el codec adecuado.")


```


# Calcula la media de los precios de cierre

```python

media_cierre = df['Cierre'].mean()
```

```python

columna_fecha = 'Fecha' if 'Fecha' in df.columns else df.columns[0]
```


# Crea un gráfico de líneas

```python

plt.figure(figsize=(10, 6))
plt.plot(df[columna_fecha], df['Cierre'], label='Precio de Cierre')
plt.axhline(y=media_cierre, color='r', linestyle='--', label='Media de Cierre')
plt.title('Evolución del Precio de Cierre y Media')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
